## 2 Розподіл Бернуллі

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bernoulli.html).

````python
from scipy.stats import bernoulli
````

````python
p = 0.3
````

### 2.1 Щільність

````python
x = np.arange(-0.5, 2, 0.5)
````

Виведіть значення функції щільності розподілу Бернуллі з параметром $p$ в точках вектора $x$.

````python
y = bernoulli.pmf(x, p)
print(y)
````

Зобразіть графічно значення функції щільності розподілу Бернуллі з параметром $p$ в точках вектора $x$.

````python
plt.stem(x, y)
plt.show()
````

### 2.2 Функція розподілу

Виведіть значення функції розподілу розподілу Бернуллі з параметром $p$ в точках вектора $x$.

````python
y = bernoulli.cdf(x, p)
print(y)
````

Зобразіть графічно значення функції розподілу розподілу Бернуллі з параметром $p$ в точках вектора $x$.

````python
plt.plot(x, y, drawstyle='steps-post')
plt.show()
````

### 2.3 Випадкові величини

Згенеруйте $100$ випадкових величин з розподілу Бернуллі з параметром $p$.

````python
bern_rvs = bernoulli.rvs(p=p, size=100)
print(bern_rvs)
````

Виведіть кількість кожного елемента в отриманому випадковому векторі.

````python
print('Кількість нулів : {}'.format(sum(bern_rvs==0)))
print('Кількість одиниць : {}'.format(sum(bern_rvs==1)))
````

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

````python
plt.hist(bern_rvs)
plt.show()
````

## 3 Біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html).

````python
from scipy.stats import binom
````

````python
n = 10
p = 0.4
````

### 3.1 Щільність

````python
x = np.arange(-0.5, n+1, 0.5)
````

Виведіть значення функції щільності біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

````python
# Місце для Вашого коду

y = binom.pmf(x, n, p)
print(y)
````

Зобразіть графічно значення функції щільності біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

````python
# Місце для Вашого коду

fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()

````

Зобразіть графічно значення функції щільності біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

````python
x1 = np.arange(0, n+1)
````

````python
# Місце для Вашого коду

y = binom.pmf(x1, n, p)
print(y)

fig, ax = plt.subplots(1, 1)
ax.plot(x1, binom.pmf(x1, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x1, 0, binom.pmf(x1, n, p), colors='b', lw=5, alpha=0.5)
rv = binom(n, p)
ax.vlines(x1, 0, rv.pmf(x1), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
````

### 3.2 Функція розподілу

Виведіть значення функції розподілу біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

````python
# Місце для Вашого коду

y = binom.cdf(x, n, p)
print(y)
````

Зобразіть графічно значення функції розподілу біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

````python
# Місце для Вашого коду

fig, ax = plt.subplots(1, 1)
ax.plot(x, y, 'bo', ms=8)
ax.vlines(x, 0, y, colors='b', lw=5, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.cdf(x), colors='k', linestyles='-', lw=1)
ax.legend(loc='best', frameon=False)
plt.show()
````

### 3.3 Випадкові величини

Згенеруйте $100$ випадкових величин з біноміального розподілу з параметрами $n, p$.

````python
# Місце для Вашого коду

a = binom.rvs(n, p, size=100)
print(a)
````

Виведіть кількість кожного елемента в отриманому випадковому векторі.

````python
# Місце для Вашого коду
for i in range(n+1):
    print("К-сть {} : {}".format(i, sum(a==i)))
````

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

````python
# Місце для Вашого коду

plt.hist(a)
plt.show()
````

## 5 Розподіл Пуассона

````python
from scipy.stats import poisson
````

````python
mu = 0.6
````

````python
x = np.arange(-0.5, 11, 0.5)
````

Зобразіть графічно значення функції щільності розподілу Пуассона з параметром $\mu$ в точках вектора $x1$.

````python
x1 = np.arange(0, 11)
````

````python
# Місце для Вашого коду

y = poisson.pmf(x1, mu)
print(y)

fig, ax = plt.subplots(1, 1)
ax.plot(x1, y, 'bo', ms=8)
ax.vlines(x1, 0, y, colors='b', lw=5, alpha=0.5)
ax.legend(loc='best', frameon=False)
plt.show()
````

### 5.2 Функція розподілу
